```python
import numpy as np
import pandas as pd
filename='F:/BDCI竞赛/剧本角色情感识别/train_dataset.tsv'#文件地址
train_data=[]
with open(filename,'r',encoding='utf-8-sig') as file_to_read:
     for line in file_to_read:
        train_data.append(list(line.strip().split()))
        

train_data=pd.DataFrame(train_data)      
```

```python
#替换空值所在的行
train_data[3] = train_data[3].fillna(value='0,0,0,0,0,0')
train_data[2] = train_data[2].fillna(value='no character')
```

```python
#设置列索引
train_data.columns=train_data.iloc[0].values#将第一行的value作为train_data列表的索引值
train_data=train_data.drop(0)#删除第一行数据
train_data=train_data.reset_index(drop=True)#重置索引为真
```

```python
#存入excel中
train_data.to_excel('F:/BDCI竞赛/剧本角色情感识别/train_data.xlsx')
```
